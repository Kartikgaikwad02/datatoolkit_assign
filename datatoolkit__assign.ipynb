```python
#3 Explain the following termsg
The difference in npYarray, npYasarray and npYasanyarrayX
The difference between Deep copy and shallow copyX

Difference between np.array(), np.asarray(), and np.asanyarray()
np.array()
Purpose: It creates a new NumPy array from any object (like a list, tuple, or another array). It will always make a copy of the data unless explicitly told not to (via the copy=False argument).
Key point: Even if the input is already a NumPy array, np.array() will create a copy unless you specify copy=False.

np.asarray()
Purpose: This function converts the input to a NumPy array, but it won't copy the data if the input is already a NumPy array. If the input is not a NumPy array (e.g., a list or tuple), it will convert it to one.
Key point: It only creates a new array if the input is not already a NumPy array, avoiding unnecessary copies.

np.asanyarray()
Purpose: Similar to np.asarray(), but with a key difference—it preserves subclass types like matrix or masked_array. If the input is already an array or subclass of an array, np.asanyarray() won't convert it to a basic NumPy array.
Key point: It will allow the input to remain as a subclass of ndarray (like matrix or masked_array), unlike np.asarray() which always converts to a basic ndarray.


2. Difference Between Deep Copy and Shallow Copy
Shallow Copy
Definition: A shallow copy creates a new object, but does not recursively copy the objects contained within the original. It copies the references (pointers) to the original objects, not the objects themselves.
Key point: Changes made to the objects within the copied array will affect the original array because both arrays share the same references to the inner objects.

Deep Copy
Definition: A deep copy creates a completely independent copy of the original object and all the objects contained within it. The copy is fully detached from the original, so changes in the copied object do not affect the original one.
Key point: Both the outer object and all inner objects are fully copied, meaning changes in one will not reflect in the other.
```

```python
#5 Create a NumPy array with random integers between 1 and 10 of shape (5, 6)) After creating the array perform the following operations:
 #a)Extract all even integers from array.
 #b)Extract all odd integers from array

import numpy as np

# Step 1: Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, (5, 6))

# Step 2a: Extract all even integers from the array
even_integers = array[array % 2 == 0]

# Step 2b: Extract all odd integers from the array
odd_integers = array[array % 2 != 0]

# Display the original array and the results
print("Original array:")
print(array)

print("\nEven integers:")
print(even_integers)

print("\nOdd integers:")
print(odd_integers)
```

```python
#Create two NumPy arrays, x and y, each containing 100 random float values between 0 and 1. Perform the following tasks using Matplotlib and NumPy:
#a) Create a scatter plot using x and y, setting the color of the points to red and the marker style to 'o'.
#b) Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'.
#c) Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'.
#d) Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'.
#e) Set the title of the plot as 'Advanced Scatter Plot of Random Values'.
#f) Display a legend for the scatter plot, the horizontal line, and the vertical line.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two NumPy arrays, x and y, with 100 random float values between 0 and 1
x = np.random.rand(100)
y = np.random.rand(100)

# Step 2a: Create a scatter plot using x and y, set the color to red and marker style to 'o'
plt.scatter(x, y, color='red', marker='o', label='Random Scatter')

# Step 2b: Add a horizontal line at y = 0.5 using a dashed line style and label it as 'y = 0.5'
plt.axhline(y=0.5, color='blue', linestyle='--', label='y = 0.5')

# Step 2c: Add a vertical line at x = 0.5 using a dotted line style and label it as 'x = 0.5'
plt.axvline(x=0.5, color='green', linestyle=':', label='x = 0.5')

# Step 2d: Label the x-axis as 'X-axis' and the y-axis as 'Y-axis'
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Step 2e: Set the title of the plot as 'Advanced Scatter Plot of Random Values'
plt.title('Advanced Scatter Plot of Random Values')

# Step 2f: Display a legend for the scatter plot, the horizontal line, and the vertical line
plt.legend()

# Show the plot
plt.show()
```

```python
#Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and  Perform the following tasks using Matplotlib:
#a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
#b) Label the x-axis as 'Date'.
#c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a time-series dataset in a DataFrame
date_range = pd.date_range(start='2023-01-01', periods=30, freq='D')  # 30 days of data
temperature = np.random.uniform(15, 30, 30)  # Random temperatures between 15 and 30 degrees Celsius
humidity = np.random.uniform(40, 80, 30)  # Random humidity between 40% and 80%

# Creating the DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})

# Step 2: Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.set_xlabel('Date')  # Label for x-axis
ax1.set_ylabel('Temperature (°C)', color='tab:red')  # Label for the left y-axis
ax1.plot(data['Date'], data['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Step 3: Create another y-axis for Humidity
ax2 = ax1.twinx()  # Create a second y-axis sharing the same x-axis
ax2.set_ylabel('Humidity (%)', color='tab:blue')  # Label for the right y-axis
ax2.plot(data['Date'], data['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Step 4: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
fig.tight_layout()  # Adjust layout to make space for the labels
plt.show()
```

```python
#Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
#tasks using Matplotlib:
#a) Plot a histogram of the data with 30 bins.
#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
#c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
#d) Set the title of the plot as 'Histogram with PDF Overlay'.


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Std Dev = 1

# Step 2a: Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Data Histogram')

# Step 2b: Overlay a line plot representing the normal distribution's PDF
xmin, xmax = plt.xlim()  # Get the x-axis limits from the histogram
x = np.linspace(xmin, xmax, 100)  # Create 100 equally spaced points on the x-axis
p = norm.pdf(x, loc=0, scale=1)  # Generate the normal distribution PDF
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# Step 2c: Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Step 2d: Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Add a legend to differentiate between the histogram and the PDF
plt.legend()

# Display the plot
plt.show()
```

```python
#create a Seaborn scatter plot where points are colored based on their position relative to the origin (quadrants), we can generate two random arrays and assign colors depending on which quadrant the points fall into. Here's how to do it:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate two random arrays of 100 points each
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)  # Random points from a normal distribution
y = np.random.randn(100)

# Step 2: Determine the quadrant of each point based on x and y values
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Q1'  # First Quadrant
    elif x < 0 and y >= 0:
        return 'Q2'  # Second Quadrant
    elif x < 0 and y < 0:
        return 'Q3'  # Third Quadrant
    else:
        return 'Q4'  # Fourth Quadrant

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y})

# Step 3: Apply the function to assign quadrants
data['Quadrant'] = data.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

# Step 4: Create a scatter plot using Seaborn, coloring by 'Quadrant'
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='Set1', s=100)

# Step 5: Add axis labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Step 6: Add a legend
plt.legend(title='Quadrant')

# Show the plot
plt.show()
```

```python
#With Bokeh, plot a line chart of a sine wave function, add rid lines, label the axes, and set the title as 'SineWave Function'8
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Prepare the output to display in the notebook (if using Jupyter notebook)
output_notebook()

# Step 2: Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 100)  # X values from 0 to 4π
y = np.sin(x)  # Y values are sine of x

# Step 3: Create a Bokeh figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Step 4: Add a line plot of the sine wave to the figure
p.line(x, y, legend_label="Sine Wave", line_width=2, line_color="blue")

# Step 5: Enable grid lines (by default they are enabled, but we'll ensure they're visible)
p.xgrid.visible = True
p.ygrid.visible = True

# Step 6: Show the plot
show(p)
```

```python
#Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
#values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
#Bar Chart'8

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.transform import factor_cmap
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256

# Step 1: Prepare the output to display in the notebook (if using Jupyter notebook)
output_notebook()

# Step 2: Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']  # 5 categories
values = np.random.randint(10, 100, size=5)  # Random values for each category

# Step 3: Create a DataFrame and convert to a ColumnDataSource for Bokeh
data = pd.DataFrame({'categories': categories, 'values': values})
source = ColumnDataSource(data)

# Step 4: Create a Bokeh figure for the bar chart
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           plot_height=400, plot_width=600, toolbar_location=None)

# Step 5: Add the bar chart to the figure
p.vbar(x='categories', top='values', width=0.8, source=source,
       legend_field='categories',
       line_color='white',
       fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Step 6: Add hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Step 7: Customize the chart (remove grid lines and add legend)
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Step 8: Display the plot
show(p)
```

```python
#Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'8

import numpy as np
import plotly.graph_objects as go

# Step 1: Generate a random dataset
np.random.seed(42)  # For reproducibility
x = np.arange(0, 100)  # X values (0 to 99)
y = np.random.randn(100)  # Y values (random data from a normal distribution)

# Step 2: Create a basic line plot
fig = go.Figure()

# Add a line trace to the figure
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Step 3: Customize the plot (add labels and title)
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Step 4: Display the plot
fig.show()
```

```python
# Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'
import numpy as np
import plotly.graph_objects as go

# Step 1: Generate random data for the pie chart
np.random.seed(42)  # For reproducibility
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)  # Random values for the pie chart

# Step 2: Create an interactive pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                             hoverinfo='label+percent',
                             textinfo='label+percent',
                             textfont_size=15)])

# Step 3: Customize the layout (title)
fig.update_layout(
    title_text='Interactive Pie Chart',
)

# Step 4: Display the pie chart
fig.show()
```
